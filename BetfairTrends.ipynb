```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('./bpi-USD-close_data.csv')
```

```python
trends = pd.read_csv('./googleTrends.csv')
```

```python
combined = pd.merge(prices, trends, on='Date')
```

```python
maxMin = combined['bitcoin'].max() - combined['bitcoin'].min()
mean = combined['bitcoin'].mean()
combined['bitcoin'] = combined['bitcoin'].map(lambda x: (x - mean) / maxMin)
```

```python
maxMin = combined['Close Price'].max() - combined['Close Price'].min()
mean = combined['Close Price'].mean()
combined['Close Price'] = combined['Close Price'].map(lambda x: (x - mean) / maxMin)
```

```python
plt.plot(combined['Close Price'])
```

```python
plt.plot(combined['bitcoin'])
```

```python
plt.show()
```
